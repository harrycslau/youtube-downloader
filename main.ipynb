# YouTube to MP4/MP3 Downloader

This notebook allows you to download YouTube videos as MP4 (video) or MP3 (audio) with a quality selection option.

**Requirements:**  
- `pytube` for downloading YouTube videos  
- `moviepy` for MP3 conversion (optional, for audio extraction)

```python
# Install required libraries
!pip install pytube moviepy
```

```python
from pytube import YouTube
from moviepy.editor import AudioFileClip
import os

def download_youtube(url, download_type='mp4', quality='highest'):
    yt = YouTube(url)
    if download_type == 'mp4':
        if quality == 'highest':
            stream = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
        elif quality == 'lowest':
            stream = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').first()
        else:
            stream = yt.streams.filter(res=quality, progressive=True, file_extension='mp4').first()
        if stream is None:
            print("Requested quality not available. Downloading highest quality.")
            stream = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
        out_file = stream.download()
        print(f"Downloaded video: {out_file}")
    elif download_type == 'mp3':
        stream = yt.streams.filter(only_audio=True).first()
        out_file = stream.download()
        base, _ = os.path.splitext(out_file)
        mp3_file = base + '.mp3'
        audio_clip = AudioFileClip(out_file)
        audio_clip.write_audiofile(mp3_file)
        audio_clip.close()
        os.remove(out_file)
        print(f"Downloaded audio: {mp3_file}")
    else:
        print("Invalid download_type. Use 'mp4' or 'mp3'.")

# Example usage:
# url = "https://www.youtube.com/watch?v=YOUR_VIDEO_ID"
# download_youtube(url, download_type='mp4', quality='720p')  # For specific resolution
# download_youtube(url, download_type='mp4', quality='highest')  # For highest quality video
# download_youtube(url, download_type='mp3')  # For audio only
```

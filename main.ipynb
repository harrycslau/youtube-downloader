# YouTube to MP4/MP3 Downloader
This notebook allows you to download YouTube videos as MP4 (video) or MP3/M4A (audio) with quality selection options.

**Requirements:**  
- `yt-dlp` - A more reliable YouTube downloader that handles API changes better
- `ffmpeg` - For audio/video processing (automatically handled by yt-dlp)

```python
# filepath: /Users/harrycslau/Documents/Git Repositories/youtube-downloader/main.ipynb
import yt_dlp
import os
from pathlib import Path

def download_youtube(url, download_type='mp4', quality='best', output_path='./downloads'):
    """
    Download YouTube videos as MP4 or MP3/M4A
    
    Args:
        url (str): YouTube URL
        download_type (str): 'mp4' for video, 'mp3' for audio MP3, 'm4a' for audio M4A
        quality (str): Quality options:
            - For MP4: 'best', 'worst', '720p', '480p', '360p', '1080p'
            - For MP3/M4A: 'best', 'worst' (audio quality)
        output_path (str): Directory to save downloads
    """
    
    # Create output directory if it doesn't exist
    Path(output_path).mkdir(parents=True, exist_ok=True)
    
    if download_type == 'mp4':
        # Video download options
        if quality == 'best':
            format_selector = 'best[ext=mp4]'
        elif quality == 'worst':
            format_selector = 'worst[ext=mp4]'
        elif quality in ['720p', '480p', '360p', '1080p']:
            height = quality.replace('p', '')
            format_selector = f'best[height<={height}][ext=mp4]/best[ext=mp4]'
        else:
            format_selector = 'best[ext=mp4]'
            
        ydl_opts = {
            'format': format_selector,
            'outtmpl': f'{output_path}/%(title)s.%(ext)s',
        }
        
    elif download_type == 'mp3':
        # Audio download options for MP3
        ydl_opts = {
            'format': 'bestaudio/best' if quality == 'best' else 'worstaudio/worst',
            'outtmpl': f'{output_path}/%(title)s.%(ext)s',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192' if quality == 'best' else '128',
            }],
        }
    elif download_type == 'm4a':
        # Audio download options for M4A
        ydl_opts = {
            'format': 'bestaudio/best' if quality == 'best' else 'worstaudio/worst',
            'outtmpl': f'{output_path}/%(title)s.%(ext)s',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'm4a',
                'preferredquality': '192' if quality == 'best' else '128',
            }],
        }
    else:
        print("Invalid download_type. Use 'mp4', 'mp3', or 'm4a'.")
        return
    
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            # Get video info first
            info = ydl.extract_info(url, download=False)
            title = info.get('title', 'Unknown')
            duration = info.get('duration', 0)
            
            print(f"Title: {title}")
            print(f"Duration: {duration//60}:{duration%60:02d}")
            print(f"Downloading as {download_type.upper()} with {quality} quality...")
            
            # Download the video/audio
            ydl.download([url])
            print(f"✅ Download completed successfully!")
            
    except Exception as e:
        print(f"❌ Error occurred: {e}")
        print("This might be due to:")
        print("1. Invalid URL")
        print("2. Video is private or restricted")
        print("3. Network connection issues")

def list_available_formats(url):
    """List all available formats for a YouTube video"""
    try:
        ydl_opts = {'listformats': True}
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.extract_info(url, download=False)
    except Exception as e:
        print(f"Error: {e}")
```

```python
# Example usage (replace with your YouTube URL)
url = "https://youtu.be/xxxxxxxxxxxx"

# List all available formats (uncomment to see options)
list_available_formats(url)
```
